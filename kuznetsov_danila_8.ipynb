```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import numpy as np
from sklearn.cluster import KMeans


from sklearn.datasets import load_digits
digits = load_digits()
digits.data.shape
```

```python
kmeans = KMeans(n_clusters=10, random_state= 0)
cluster = kmeans.fit_predict(digits.data)
kmeans.cluster_centers_.shape
```

```python
fig, ax = plt.subplots(2, 5, figsize=(8, 3))
centers = kmeans.cluster_centers_.reshape(10, 8, 8)
for axi, center in zip(ax.flat, centers):
    axi.set(xticks=[], yticks=[])
    axi.imshow(center, interpolation='nearest', cmap=plt.cm.binary)
```

```python
X1 = np.array([[5,3],
               [10,15],
               [15,12],
               [24,10],
               [30,45],
               [85,70],
               [71,80],
               [60,78],
               [55,52],
               [80,91]])

plt.scatter(X1[:, 0], X1[:, 1], s = 50);

plt.show()
```

```python
# 2 clusters

kmeans = KMeans(n_clusters=2)
kmeans.fit(X1)
y_kmeanss = kmeans.predict(X1)

plt.scatter(X1[:, 0], X1[:, 1], c = y_kmeanss, s = 20, cmap = 'summer')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', s = 100, alpha= 0.9);
plt.show()
```

```python
# 3 clusters

kmeans = KMeans(n_clusters=3)
kmeans.fit(X1)
y_kmeanss = kmeans.predict(X1)

plt.scatter(X1[:, 0], X1[:, 1], c = y_kmeanss, s = 20, cmap = 'summer')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', s = 100, alpha= 0.9);
plt.show()
```

```python
# 4 clusters

kmeans = KMeans(n_clusters=4)
kmeans.fit(X1)
y_kmeanss = kmeans.predict(X1)

plt.scatter(X1[:, 0], X1[:, 1], c = y_kmeanss, s = 20, cmap = 'summer')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c = 'blue', s = 100, alpha= 0.9);
plt.show()
```

```python
iris.drop(columns=['species'])
iris.head()
```

```python
X2 = iris.loc[:, iris.columns!='species'].to_numpy()

plt.scatter(X2[:, 0], X2[:, 1], s = 20);

plt.show()
```
